* it enclosed in ()
* used to create the multiple variables with single statement
* support only 2 built-in methods

```python
fruits=('apple','cherry','dates')
```

```python
fruits
```

```python
len(fruits)
```

```python
fruits[0]
```

```python
fruits[-1]
```

```python
'cherry' in fruits
```

```python
fruits[0]='pineapple'
```

##Dictionary

a dictionary is an unodered collection of items

* value stored in key-value pair
* keys are used to retrive the data from dict
* are used to store the pieces of information eg. details about person
* enclosed in {}

```python
student={
    'name':'sush',
    'age':'21',
    'grade':'A'

}
```

```python
student
```

```python
type(student)
```

```python
student2
```

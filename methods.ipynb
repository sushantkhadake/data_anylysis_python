###.format()
used to combine values of another data type
eg.int,float,bool,list etc

used for maney combinations of data

```python
#inputs
cost_of_ice_bag=1.25
profit_margin=0.2
number_of_bags=500

#template for the output message
output_template='''If a grocery store sells ice bags at $ {} per bag,with a profit margin of {}%,
then the total profit it makes by selling {} ice bags is $ {}'''

print(output_template)
```

```python
#inserting values into the string
total_profit=cost_of_ice_bag*profit_margin*number_of_bags

output_message=output_template.format(cost_of_ice_bag,profit_margin*100,number_of_bags,total_profit)

print(output_message)
```

reverse the order in a list

```python
#declare the list
list2=[1,'abc','science',35,'math']

#for loop
for i in range(len(list2)-1,-1,-1):
  print(list2[i])
```

```python
my_list_slice = [1, 'abc', 'science', 35, 'math']
print(f"Original list: {my_list_slice}\n")


##using Slice But it create a new list
reversed_list_slice = my_list_slice[::-1]
print(f"New reversed list : {reversed_list_slice}")
```

```python
words = ["banana", "apple", "mango", "cherry"]

for i in range(len(words)):
    for j in range(len(words)-1):
        if words[j] > words[j+1]:
            temp = words[j]
            words[j] = words[j+1]
            words[j+1] = temp

print(words)
```
